```python
import pandas as pd

students = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Aadesh", "Rahul", "Sneha"]
})

marks = pd.DataFrame({
    "ID": [1, 2, 4],
    "Marks": [85, 90, 75]
})

print("Students DataFrame:")
print(students)

print("\nMarks DataFrame:")
print(marks)
```

```python
# Performing an inner join on the "ID" column
inner_join = pd.merge(students, marks, on="ID", how="inner")

print("Inner Join:")
print(inner_join)
```

```python
# Performing a left join on the "ID" column
left_join = pd.merge(students, marks, on="ID", how="left")

print("Left Join:")
print(left_join)
```

```python
# Performing a Outer join on the "ID" column
outer_join = pd.merge(students, marks, on="ID", how="outer")

print("Outer Join:")
print(outer_join)
```

```python
# Performing a Right join on the "ID" column
right_join = pd.merge(students, marks, on="ID", how="right")

print("Right Join:")
print(right_join)
```
