```python
# Real-World Example (Student + Scores)

import pandas as pd

student_info = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Aadesh", "Rahul", "Sneha"]
})

exam_scores = pd.DataFrame({
    "ID": [1, 2, 3],
    "Score": [88, 92, 85]
})

combined = pd.merge(student_info, exam_scores, on="ID")

print("Combined Data:")
print(combined)
```
